```python
import csv
import matplotlib.pyplot as plt

employees = []

with open(r"C:\Users\User\Desktop\employee.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        emp = {
            "empid": row["empid"],
            "ename": row["ename"],
            "designation": row["designation"],
            "salary": int(row["salary"])
        }
        employees.append(emp)

print("Total number of employees:", len(employees))
```

```python
total_salary = 0

for emp in employees:
    total_salary += emp["salary"]

average_salary = total_salary / len(employees)
print("Average Salary of Employees:", average_salary)
```

```python
highest_paid = employees[0]

for emp in employees:
    if emp["salary"] > highest_paid["salary"]:
        highest_paid = emp

print("Highest Paid Employee Details:")
print(highest_paid)
```

```python
print("Employees with Salary above 30000:")

for emp in employees:
    if emp["salary"] > 30000:
        print(emp)
```

```python
employees_sorted = sorted(employees, key=lambda x: x["salary"], reverse=True)

print("Top 3 Highest Paid Employees:")
for emp in employees_sorted[:3]:
    print(emp)
```

```python
names = []
salaries = []

for emp in employees[:10]:   # first 10 for clarity
    names.append(emp["ename"])
    salaries.append(emp["salary"])

plt.bar(names, salaries)
plt.xlabel("Employee Name")
plt.ylabel("Salary")
plt.title("Employee Names vs Salaries")
plt.xticks(rotation=45)
plt.show()
```

```python
designation_salary = {}

for emp in employees:
    designation_salary[emp["designation"]] = \
        designation_salary.get(emp["designation"], 0) + emp["salary"]

plt.pie(designation_salary.values(),
        labels=designation_salary.keys(),
        autopct="%1.1f%%")
plt.title("Salary Share by Designation")
plt.show()
```
